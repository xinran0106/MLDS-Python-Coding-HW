```python
## Creating a dataframe with Name, Graduation Year, Current Role and Current Company
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initializing a list for all alumni data
alumni_data = []

base_url = 'https://www.mccormick.northwestern.edu/machine-learning-data-science/people/alumni/'

for year in range(2013, 2023):
    
    url = f"{base_url}class-of-{year}.html"
    response = requests.get(url)

    if response.ok:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        
        #Class 'faculty cf'
        alumni_elements = soup.find_all(class_='faculty cf')
        
        # Iterating through each alumni 
        for alumni in alumni_elements:
            name = alumni.h3.get_text(strip=True) if alumni.h3 else 'N/A'
            grad_year = alumni.find('em').get_text(strip=True) if alumni.find('em') else f'Class of {year}'
            role_and_company = alumni.find('div', class_='student-alumni-details').get_text(strip=True) if alumni.find('div', class_='student-alumni-details') else 'N/A'
            
            if ' at ' in role_and_company:
                current_role, current_company = role_and_company.split(' at ', 1)
            else:
                current_role = role_and_company
                current_company = 'N/A'
            
            # Appending all the information to the alumni_data list
            alumni_data.append({
                'Name': name,
                'Graduation Year': grad_year,
                'Current Role': current_role,
                'Current Company': current_company
            })
    else:
        print(f"No data {year}. Status code: {response.status_code}")

# Df with alumni data
alumni_df = pd.DataFrame(alumni_data)

print(alumni_df)
alumni_df.to_csv('alumni_data.csv', index=False)
```

```python
## Top 5 Companies that alumni are working at
```

```python
alumni_df = pd.read_csv('alumni_data.csv')

# Counting the occurrences of each company 
company_counts = alumni_df['Current Company'].value_counts()

top_5_companies = company_counts.head(5)

# Print the top 5 companies
print(top_5_companies)
```

```python
##Insight 1: Roles Distribution
```

```python
alumni_df = pd.read_csv('alumni_data.csv')

# Analyzing the distribution of job titles
role_counts = alumni_df['Current Role'].value_counts().head(10)  

plt.figure(figsize=(10, 8))
role_counts.plot(kind='barh')
plt.title('Top 10 Alumni Job Titles')
plt.xlabel('Number of Alumni')
plt.ylabel('Job Titles')
plt.gca().invert_yaxis()  
plt.show()
```

```python
## Insight 2: Graduation Year and Employment Trends
```

```python
alumni_df = pd.read_csv('alumni_data.csv')

# Grouping by graduation year and counting the number of alumni employed
employment_trend = alumni_df.groupby('Graduation Year')['Name'].count()

plt.figure(figsize=(12, 6))
employment_trend.plot(kind='line', marker='o')
plt.title('Employment Trend by Graduation Year')
plt.xlabel('Graduation Year')
plt.ylabel('Number of Employed Alumni')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
