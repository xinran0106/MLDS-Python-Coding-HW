### 1. Find the number of movies that were released in theatres each year.
1. There are some erroneous values in the data
2. Visualize the number of movies each year and describe what you see
3. Explain what assumptions and fixes you plan to make in the data as you move forward with your analysis

```python
# Extract the release year
df['release_year'] = pd.to_datetime(df['in_theaters_date']).dt.year
print("There are", len(df[pd.notnull(df['release_year'])]), "movies released in theatres")

# Count the number of movies each year
movies_per_year = df['release_year'].value_counts().sort_index()

# Visualize the data
plt.figure(figsize=(10, 6))
plt.bar(movies_per_year.index, movies_per_year.values)
plt.title('Number of Movies Released in Theaters Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```

In the visualization, the number of released movies increase year by year. However, a sudden decline is observed from the year 2020 onwards. Additionally, there are instances where movies appear to have been released beyond the current year (2023).

As a corrective measure, I intend to treat movies with a release date after 2019 as anomalous data. Specifically, I will update their 'in_theaters_date' values to NaT.

```python
# Identify movies with a release date after 2019
anomalous_movies = df[df['release_year'] >= 2020]

# Update 'in_theaters_date' values to NaT for anomalous movies
df.loc[anomalous_movies.index, 'in_theaters_date'] = pd.NaT

# Visualize the corrected data
df['release_year'] = pd.to_datetime(df['in_theaters_date']).dt.year
print("There are", len(df[pd.notnull(df['release_year'])]), "movies released in theatres")
movies_per_year = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(movies_per_year.index, movies_per_year.values)
plt.title('Number of Movies Released in Theaters Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```

## Feature Engineering

### 7. Split the data into a training and test set, with the training data including movies released in theatres before 2010 and the test data including movies released in theatres in 2010 and after.

```python
# Convert 'in_theaters_date' to datetime
df['in_theaters_date'] = pd.to_datetime(df['in_theaters_date'])

# Train test split
train_data = df[df['in_theaters_date'] < '2010-01-01']
test_data = df[df['in_theaters_date'] >= '2010-01-01']
```

### 8. If your goal is to predict the critic rating before the first critic or audience rating gets posted for a movie, which columns in the data should you NOT use to create features? Update your training and test data sets to NOT include these columns.

We should not use audience_rating, audience_count, and critic_count

```python
# Columns to exclude from features
exclude_columns = ['audience_rating', 'audience_count', 'critic_count']

X_train = train_data.drop(exclude_columns, axis=1)
y_train = train_data['critic_rating']
X_test = test_data.drop(exclude_columns, axis=1)
y_test = test_data['critic_rating']

# Display the updated shapes of the training and test sets
print("Updated Training set shape:", X_train.shape, y_train.shape)
print("Updated Test set shape:", X_test.shape, y_test.shape)
```

### 9. Using only the training data, create a new DataFrame containing the following ID column and features:
1. movie_title
2. runtime_in_minutes
3. NEW: kid_friendly (1 if G or PG, 0 if other ratings)
4. NEW: dummy variable columns for each genre

```python
# Create a copy of the original DataFrame
new_features_df = X_train[['movie_title', 'runtime_in_minutes']].copy()

# Create a binary column 'kid_friendly'
new_features_df['kid_friendly'] = X_train['rating'].isin(['G', 'PG']).astype(int)

# Create dummy variables for each genre
genres = X_train['genre'].str.get_dummies(sep=', ')
new_features_df = pd.concat([new_features_df, genres], axis=1)

# Display the new DataFrame
new_features_df.head(5)
```

### 10.Create 3 new features that you think will do a good job predicting the critic_rating. Each new feature should use various combinations of the columns from your training data.

1. Release Season

```python
# Map months to seasons
season_mapping = {
    1: 'Winter',
    2: 'Winter',
    3: 'Spring',
    4: 'Spring',
    5: 'Spring',
    6: 'Summer',
    7: 'Summer',
    8: 'Summer',
    9: 'Fall',
    10: 'Fall',
    11: 'Fall',
    12: 'Winter'
}
X_train['release_season'] = X_train['in_theaters_date'].dt.month.map(season_mapping)

# Create dummy variables for seasons
season_dummies = X_train['release_season'].str.get_dummies()

# Concatenate the dummy variables to the DataFrame
new_features_df = pd.concat([new_features_df, season_dummies], axis=1)
```

2. Popularity of the genre

```python
# Create a stacked DataFrame with one row for each genre of each movie
stacked_data = train_data['genre'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('genre')
stacked_data = train_data.drop('genre', axis=1).join(stacked_data)

# Create a dictionary mapping genres to average audience count
genre_popularity_dict = stacked_data.groupby('genre')['audience_count'].mean().to_dict()

# Map the average audience count to the 'genre' column to create the new feature
X_train['genre_popularity_index'] = X_train['genre'].apply(lambda x: np.mean([genre_popularity_dict.get(genre, 0) for genre in  str(x).split(', ')]))
new_features_df['genre_popularity_index'] = X_train['genre_popularity_index']
```

3. Director's number of sucessful movies 

```python
# Define a threshold for a successful critic rating 
successful_rating_threshold = 75

# Create a dictionary to store the number of successful movies for each director
director_success_count_dict = {}

# Iterate through each row in the training data
for index, row in X_train.iterrows():
    # Extract the list of directors for the movie
    directors_list = str(row['directors']).split(', ') if not pd.isna(row['directors']) else []
    
    # Check if any director in the list has a previous successful movie
    if any(director_success_count_dict.get(director, 0) > 0 for director in directors_list):
        # If yes, increment the count for each director in the list
        for director in directors_list:
            director_success_count_dict[director] = director_success_count_dict.get(director, 0) + 1
    else:
        # If not, check if the current movie is successful
        if row['critic_rating'] >= successful_rating_threshold:
            # If successful, set the count for each director in the list to 1
            for director in directors_list:
                director_success_count_dict[director] = 1
        else:
            # If not successful, set the count for each director in the list to 0
            for director in directors_list:
                director_success_count_dict[director] = 0

# Map the Director's Success Count to the 'directors' column to create the new feature
X_train['director_success_count'] = X_train['directors'].apply(lambda x: np.mean([director_success_count_dict.get(director, 0) for director in str(x).split(', ')]))
new_features_df['director_success_count'] = X_train['director_success_count']
```

### 11.Make sure you apply the same transformations on your X_test and y_test data sets that you applied on the X_train and y_train data sets.


```python
X_train = new_features_df.drop(['movie_title'], axis=1) 

# Transformation on X_test
new_features_df_test = X_test[['movie_title', 'runtime_in_minutes']].copy()

new_features_df_test['kid_friendly'] = X_test['rating'].isin(['G', 'PG']).astype(int)

genres = X_test['genre'].str.get_dummies(sep=', ')
new_features_df_test = pd.concat([new_features_df_test, genres], axis=1)

X_test['release_season'] = X_test['in_theaters_date'].dt.month.map(season_mapping)
season_dummies_test = X_test['release_season'].str.get_dummies()
new_features_df_test = pd.concat([new_features_df_test, season_dummies_test], axis=1)

new_features_df_test['genre_popularity_index'] = X_test['genre'].apply(lambda x: np.mean([genre_popularity_dict.get(genre, 0) for genre in  str(x).split(', ')]))

new_features_df_test['director_success_count'] = X_test['directors'].apply(lambda x: np.mean([director_success_count_dict.get(director, 0) for director in str(x).split(', ')]))

X_test = new_features_df_test.drop(['movie_title'], axis=1) 
```

### 13.Fit 3 linear regression models on the training data:
1. Model 1: Use only runtime_in_minutes
2. Model 2: Use runtime_in_minutes and kid_friendly
3. Model 3: Use runtime_in_minutes, kid_friendly and the dummy columns for the genres

```python
# Model 1: Use only runtime_in_minutes
X_model1 = X_train[['runtime_in_minutes']].copy()
model1 = LinearRegression()
model1.fit(X_model1, y_train)

# Model 2: Use runtime_in_minutes and kid_friendly
X_model2 = X_train[['runtime_in_minutes', 'kid_friendly']].copy()
model2 = LinearRegression()
model2.fit(X_model2, y_train)

# Model 3: Use runtime_in_minutes, kid_friendly, and dummy columns for genres
X_model3 = X_train.drop(["Fall","Spring","Summer","Winter","genre_popularity_index","director_success_count"], axis=1)
model3 = LinearRegression()
model3.fit(X_model3, y_train)
```

### 14. Score the linear regression models on the test data by writing a function where you can input the y_test and y_pred values (y_pred = predicted values after you apply the fitted model to your X_test data), and it outputs the following metrics: R2, MAE and RMSE. Apply the function to the three models that you’ve fit so far

```python
def evaluate_regression_model(y_test, y_pred):
    # evalustion using r2, mae, and 
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    print(f"R2: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")


# Model 1 evaluation
X_test_model1 = X_test[['runtime_in_minutes']].copy()
y_pred_model1 = model1.predict(X_test_model1)
print("Model 1 Evaluation:")
evaluate_regression_model(y_test, y_pred_model1)
print()

# Model 2 evaluation
X_test_model2 = X_test[['runtime_in_minutes', 'kid_friendly']].copy()
y_pred_model2 = model2.predict(X_test_model2)
print("Model 2 Evaluation:")
evaluate_regression_model(y_test, y_pred_model2)
print()

# Model 3 evaluation
X_test_model3= X_test.drop(["Fall","Spring","Summer","Winter","genre_popularity_index","director_success_count"], axis=1)
y_pred_model3 = model3.predict(X_test_model3)
print("Model 3 Evaluation:")
evaluate_regression_model(y_test, y_pred_model3)
```

### 15.Which model performs the best so far? Which features seem to do a good job predicting the critic rating (hint: you can check p-values using statsmodels)?

Model 3 performs best so far

```python
# Add a constant term to the features for statsmodels
X_model3 = sm.add_constant(X_model3)

# Fit the models using statsmodels
ols_model3 = sm.OLS(y_train, X_model3).fit()

print("Model 3 Summary:")
print(ols_model3.summary())
```

Except for some of the dummy columns for genre, the features all do a good job predicting the critic rating with P-value < 0.05
